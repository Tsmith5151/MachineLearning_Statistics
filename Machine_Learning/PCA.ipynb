```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data/wine.csv',header=None)

data.columns = ['Class label', 'Alcohol', 'Malic acid', 'Ash', 
'Alcalinity of ash', 'Magnesium', 'Total phenols', 
'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 
'Color intensity', 'Hue', 'OD280/OD315 of Diluted Wines', 'Proline']
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

X = data.ix[:,1:]
y = data.ix[:,0]

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.70,random_state=123)
```

```python
from sklearn.preprocessing import StandardScaler

X_train_std = StandardScaler().fit_transform(X_train)
X_test_std = StandardScaler().fit_transform(X_test)
```

**Principal Component Analysis**

```python
mat = np.cov(X_train_std.T)

#Eigen Values and Vector
eig_vals, eig_vect = np.linalg.eig(mat)

var = sum(eig_vals)
var_exp = [(i / var) for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
```

```python
fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(111)
ax2 = ax.twinx()
ax2.step(range(1,14),cum_var_exp,color='g',lw=2.0)
ax.plot(range(1,14),var_exp,color='r',lw=2.0)
```

```python
from sklearn.decomposition import PCA
pc = PCA(n_components=2).fit(X_train_std)
proj = pc.transform(X_train_std)
```

```python
plt.scatter(proj[:,0],proj[:,1],c=y_train,cmap='cool')
plt.ylabel('PC2')
plt.xlabel('PC1')
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr = lr.fit(proj, y_train)
```

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X,y,classifier,grid_res = 0.02):
    
    #Decision Surface
    x1_min, x1_max = X[:, 0].min() - .50, X[:, 0].max() + .50
    x2_min, x2_max = X[:, 1].min() - .50, X[:, 1].max() + .50
    
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, grid_res),
                           np.arange(x2_min, x2_max, grid_res))
    
    
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    
    #Make Predictions given first 2 PCs
    print(np.array([xx1.ravel(), xx2.ravel()]).T.shape)
    
    Z = Z.reshape(xx1.shape)

    cm = plt.cm.RdBu
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])
    
    #Countour Plot
    plt.contourf(xx1,xx2,Z,alpha=0.50,cmap='jet')
    
    plt.xlim(x1_min,x1_max)
    plt.ylim(x2_min,x2_max)
    
    #Plot Datapoints
    plt.scatter(X[:,0],X[:,1],c=y,cmap='jet')
    
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend(loc='lower left')
    plt.tight_layout()
```

```python
plot_decision_regions(proj, y_train,classifier=lr)
```

```python
from sklearn.decomposition import KernelPCA
pc = KernelPCA(n_components=2,gamma=20,kernel='linear').fit(X_train_std)
proj = pc.transform(X_train_std)

plt.scatter(proj[:,0],proj[:,1],c=y_train,cmap='cool')
plt.ylabel('PC2')
plt.xlabel('PC1')
```

In the code below, we will look at implementing data reduction techniques for the MNIST dataset. This dataset contains labels for each image, but we will only be approaching the dataset from an unsupervised manner and will not include the labels for the analysis.
From TensorFlow, the MNIST dataset can be imported by running the following code:

```python
import tensorflow
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("data/MNIST_data/", one_hot=True)
```

The MNIST data is split into three parts:
55,000 data points of training data (mnist.train)
10,000 points of test data (mnist.test)
5,000 points of validation data (mnist.validation).
Each image is 28 pixels by 28 pixels. The images are flatten out into a vector of 28x28 = 784 numbers. Also, each entry in the tensor is a pixel intensity between 0 and 1, for a particular pixel in a particular image.

```python
def process_data():
    train = mnist.train.images
    test = mnist.test.images
    print('Original Training Shape:',train.shape)
    
    l = pd.DataFrame(mnist.train.labels)
    labels = l.idxmax(axis=1)
    print('Labels: ',labels.shape[0])
    
    #Reshape Data into 28x28 pixels
    train_r = train.reshape(train.shape[0],28,28)
    test_r = test.reshape(test.shape[0],28,28)
    
    print('Reformat to 28x28 Pixels Shape:',train_r.shape)
    return(train,test,train_r,test_r,labels)

train,test,train_r,test_r,labels = process_data()
```

```python
def plt_gallery(x,n_col=4,n_row=2):
    
    "Returns Photo Gallery of Images"
    
    fig = plt.figure(figsize=(4.0*n_col,2.0*n_row))
    for img,i in enumerate(range(n_row*n_col)):
        plt.subplot(n_row,n_col,i+1)
        plt.imshow(x[img],cmap='gray',interpolation=None)
        plt.xticks(())
        plt.yticks(())

plt_gallery(train_r)
```

```python
from sklearn.decomposition import TruncatedSVD

def do_truncatedSVD(data,threshold=0.60):
        
    print('Input Dimensions: {}'.format(data.shape[1]))
    
    #Setup Truncated SVD Loop
    n_components = 2
    ex_var = 0
    print('** Performing Truncated SVD... **')
    
    while ex_var < threshold:
        model = TruncatedSVD(n_components=n_components)
        pc = model.fit_transform(data)
        ex_var = np.sum(model.explained_variance_ratio_)
        n_components += 1

    print('Total Number of Components: {}'.format(model.n_components))
    print('Total Explained Variance: {}'.format(ex_var))
    print('Reduced Dimensions: {}'.format(pc.shape))
    return(pc)
```

```python
train_pc = pd.DataFrame(do_truncatedSVD(train))
```

```python
train_pc.columns = ['PC{}'.format(i+1) for i in range(0,train_pc.shape[1])]
d = pd.concat([train_pc,labels],axis=1)
d = d.rename(columns={0:'Labels'})
```

```python
sns.set_style('ticks')
sns.lmplot('PC1','PC2', data=d, hue='Labels', fit_reg=False,size=8)
plt.title('Principal Component Analysis w/ MNIST Pixels',fontsize=18)
```
